Import neccessary libs

```python
import pandas as pd
import dateutil
```

Get dataset from store

```python
df = pd.read_pickle('./dataset/nasdaq-daily.pkl')
```

Remove unused columns and add year

```python
df = df.drop(columns=['High', 'Low', 'Volume'])
df['Year'] = df['Date'].dt.year
```

Find unique symbols in dataset and get count

```python
symbols = df.Symbol.unique()
len(symbols)
```

Define examined period

```python
periods = range(1997, 2020)
```

Define resulting dataframe

```python
df_periodical_returns = pd.DataFrame([], columns=periods, index=symbols).fillna('-')
```

Loop through stocks, fill out resulting table with return calculated from yearly first day and last day

```python
for symbol in symbols[0:10]:
    symbol_df = df.loc[df['Symbol'] == symbol].copy()
    symbol_df.reset_index(drop=True, inplace=True)
    for p in periods:
        df_period = symbol_df.loc[symbol_df['Year'] == p]
        if(len(df_period.index) == 0):
            continue
        first_day = df_period.iloc[0]
        last_day = df_period.iloc[len(df_period)-1]
        df_periodical_returns.at[symbol, p] = (last_day['Close'] - first_day['Open']) / first_day['Open']
```
